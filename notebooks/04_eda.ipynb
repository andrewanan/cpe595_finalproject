```python
import pandas as pd, matplotlib.pyplot as plt, pathlib
PROC = pathlib.Path("../data/processed")
df = pd.read_csv(PROC/"btc_master.csv", parse_dates=["Date"]).set_index("Date")

# 1 – dual‑axis line plot
fig, ax1 = plt.subplots(figsize=(11,4))
df["Close"].plot(ax=ax1, label="BTC Close")
ax1.set_ylabel("Price  (USD)")

ax2 = ax1.twinx()
df["sentiment"].plot(ax=ax2, color="tab:red", alpha=.55, label="Sentiment")
ax2.set_ylabel("Mean VADER score")

fig.legend(loc="upper left")
plt.title("BTC daily price vs Twitter sentiment")
plt.show()

# 2 – simple correlation
corr = df["Close"].pct_change().corr(df["sentiment"])
print("Correlation (daily %‑change  vs  sentiment) =", round(corr, 3))
```
